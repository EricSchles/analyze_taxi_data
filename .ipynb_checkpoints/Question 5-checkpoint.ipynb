# Data preprocessing

We'll start by compressing the data to the average `Trip_distance` and average `Fare_amount` per hour over the course of the month.

```python
green = pd.read_csv("green_taxi_cleaned.csv")
green["lpep_pickup_datetime"] = pd.to_datetime(green["lpep_pickup_datetime"])
green["Lpep_dropoff_datetime"] = pd.to_datetime(green["Lpep_dropoff_datetime"])
```

```python
@jit
def generate_hour_column(x):
    x["hour"] = x["lpep_pickup_datetime"].hour
    return x
```

```python
nCores = cpu_count()
green = dd.from_pandas(green, npartitions=nCores).map_partitions(
      lambda df : df.apply(generate_hour_column, axis=1)
).compute(get=get)
```

```python
@jit
def generate_day_column(x):
    x["day"] = x["lpep_pickup_datetime"].day
    return x
```

```python
nCores = cpu_count()
green = dd.from_pandas(green, npartitions=nCores).map_partitions(
      lambda df : df.apply(generate_day_column, axis=1)
).compute(get=get)
```

```python
def get_averages(green):
    distance_average = []
    fare_average = []
    day = []
    hour = []
    for group in green.groupby(["day","hour"]):
        distance_average.append(group[1]["Trip_distance"].mean())
        fare_average.append(group[1]["Fare_amount"].mean())
        day.append(group[0][0])
        hour.append(group[0][1])
    trip = pd.DataFrame()
    trip["distance"] = np.array(distance_average)
    trip["day"] = np.array(day)
    trip["hour"] = np.array(hour)
    fare = pd.DataFrame()
    fare["amount"] = np.array(fare_average)
    fare["day"] = np.array(day)
    fare["hour"] = np.array(hour)
    return trip, fare
```

```python
trip, fare = get_averages(green)
```

```python
def create_timestamp(x):
    return datetime(2015, 9, int(x["day"]), int(x["hour"]), 0, 0)
```

```python
trip["timestamp"] = trip.apply(create_timestamp, axis=1)
trip.index = trip["timestamp"]
fare["timestamp"] = fare.apply(create_timestamp, axis=1)
fare.index = fare["timestamp"]
```

```python
trip.to_csv("trip_ave.csv", index=False)
fare.to_csv("fare_ave.csv", index=False)
```

```python
trip = pd.read_csv("trip_ave.csv")
trip["timestamp"] = pd.to_datetime(trip["timestamp"])
trip.index = trip["timestamp"]
fare = pd.read_csv("fare_ave.csv")
fare["timestamp"] = pd.to_datetime(fare["timestamp"])
fare.index = fare["timestamp"]
```

# Doing some timeseries analysis

Before we are ready to do our forecasting, we need to construct a reasonable model.  Let's start by first plotting our two data sets.

```python
trip["distance"].plot()
```

```python
fare["amount"].plot()
```

As we can see from the above two plots, our data is somewhat noisy with lots of outliers.  The reason this is a problem is our first model - ARIMA will require stationarity and invertibility of our series.

`stationarity` := A stationary process is a stochastic process whose unconditional joint probability distribution does not change when shifted in time.  As a consequence, parameters such as mean and variance do not change over time.

`invertibility` := An invertible process is a stochastic process that can be represented as a weighted sum of current and prior observations, so we can fully specify the next element in our process with infinite past and present data.

The results of these two assumptions allow us to make reliable forecasts and therefore they are essential.

One of the ways we can deal with this is by using mean or median smoothing.  In this process, we could take a sliding window and simple take the mean or the median over that sliding window.  If we make use of the median here outliers are more likely to be removed, since the median is less prone to large shifts from outliers.

Another technique we could make use of is differencing.  Since our first model will make use of ARIMA, which has builtin differencing we'll opt for that technique first.

If it doesn't produce satisfactory results, we'll first make use of rolling means or medians to smooth our data.

```python
def tsplot(y, lags=None, figsize=(10, 8), style='bmh'):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        #mpl.rcParams['font.family'] = 'Ubuntu Mono'
        layout = (3, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        qq_ax = plt.subplot2grid(layout, (2, 0))
        pp_ax = plt.subplot2grid(layout, (2, 1))
        
        y.plot(ax=ts_ax)
        ts_ax.set_title('Time Series Analysis Plots')
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.5)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.5)
        sm.qqplot(y, line='s', ax=qq_ax)
        qq_ax.set_title('QQ Plot')        
        stats.probplot(y, sparams=(y.mean(), y.std()), plot=pp_ax)

        plt.tight_layout()
    return 
```

```python
tsplot(fare["amount"])
```

```python
def acf_pacf(data):
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(211)
    fig = sm.graphics.tsa.plot_acf(data.values.squeeze(), lags=40, ax=ax1)
    ax2 = fig.add_subplot(212)
    fig = sm.graphics.tsa.plot_pacf(data, lags=40, ax=ax2)
```

```python
acf_pacf(fare["amount"])
```

```python
acf_pacf(trip["distance"])
```

These two graphs and the ones above show us that our data is neither invertible or stationary.

The autocorrelation plot corresponds with the order necessary for the moving average part of the ARIMA process.  And if the ACF doesn't converge then we fail invertibility.

Likewise, the partial autocorrelation plot corresponds with the order necessary for the autoregressive part of the ARIMA process.  And if the PACF doesn't converge then we fail stationarity.

In order to rememdy this we will difference our series using the integrated component in the ARIMA model.

```python
filtered_fare["amount"].plot()
```

As you can see the gaussian filter helps.  So filtering is definitely on the right track.  But we still aren't stationary until far too many lags in our autocorrelation or partial autocorrelation.

So we'll try some other filters, to see if we can get a slightly stronger result.

```python
filtered_fare = fare.copy()
spline = interpolate.UnivariateSpline(list(range(len(fare))), fare["amount"])
filtered_fare["amount"] = spline(list(range(len(fare))))
acf_pacf(filtered_fare["amount"])
```

```python
filtered_fare["amount"].plot()
```

Pay dirt!  We have a series that confirms for the partial autocorrelation function!!!  So we can apply an ARIMA with AR process order three and we should be able to get some predictions out!

```python
def umbrae(data, forecast):
    """
    Unscaled Mean Bounded Relative Absolute Error
    Formula taken from:
    https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5365136/
    @data - Y[i]
    @forecast - F[i]
    """
    numerator = [abs(datum - forecast[idx]) for idx, datum in enumerate(data)]
    series_one = data[1:]
    series_two = data[:-1]
    denominator = [abs(elem - series_two[idx]) for idx, elem in enumerate(series_one)]
    final_series = [numerator[idx]/(numerator[idx] + denominator[idx])
                    for idx in range(len(denominator))]
    mbrae = st.mean(final_series)
    return mbrae/(1-mbrae)
```

```python
fare_model = smt.ARIMA(filtered_fare["amount"], order=(3,0,0)).fit()
y_pred = fare_model.predict(start=10, end=len(filtered_fare), dynamic=True)
```

```python
len(y_pred), len(fare["amount"])
```

Now we'll make use of Unscaled Mean Bounded Relative Absolute Error.

At a high level this measure accounts for some of the assymmetric issues introduced by sMAPE such as scale independence and outlier resistence.

For more information on the measure, please check out the paper: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5365136/ that I list in the comment.  

```python
y = fare["amount"][9:]
umbrae(y, y_pred)
```

for comparison to see how well this model does on insample prediction, let's compare against two series - a constant series and the series itself.

```python
umbrae(y, y)
```

```python
const = [1000 for _ in range(len(y))]
umbrae(y, const)
```

Clearly higher numbers mean a worse fit, and smaller numbers are good.  So with a value of approximately 1.95 we can say our model might be "good enough" on insample prediction.  Now we are ready to do some other statistical tests to verify this claim.

First we'll look at Durbin-Watson which asks - does our model experience serial correlation?

Serial correlation means that our model's errors have a correlation with the errors in our next time period.

What this means for our model:

* unbiasedness uneffected
* consistency uneffected
* standard errors will be smaller than the true standard errors

This is basically saying our model has overfit the data so it appears to do better than it does.


```python
sm.stats.durbin_watson(fare_model.resid.values)
```

The results of the durbin_watson test confirm that our model is definitely overfitting our data.  Let's try upping the AR term to see if that results in lower overfitting.

```python
fare_model_2 = smt.ARIMA(filtered_fare["amount"], order=(4,0,0)).fit()
y_pred = fare_model_2.predict(start=10, end=len(filtered_fare), dynamic=True)
y = fare["amount"][9:]
print("Umbrae", umbrae(y, y_pred))
sm.stats.durbin_watson(fare_model_2.resid.values)
```

With a durbin_watson statistic of around 2 ~ we've gotten rid of our serial correlation!  Now let's plot the data!

```python
start = round(len(filtered_fare)/2)
# predict 2 days ahead
y_pred = fare_model_2.predict(start=start, end=len(filtered_fare)+48, dynamic=True)
pred = [np.nan for _ in range(0, start)] + list(y_pred)
pred = np.array(pred)
```

```python
def gen_timestamps(start, num_hours):
    cur_time = start
    new_timestamps = []
    for hour in range(1, num_hours):
        new_timestamp = cur_time + timedelta(hours=hour)
        new_timestamps.append(new_timestamp)
    return new_timestamps
```

```python
fare_result_df = pd.DataFrame()
new_timestamps = gen_timestamps(fare["timestamp"][-1], 50)
timestamps = list(fare["timestamp"]) + new_timestamps
timestamps = np.array(timestamps)
num_nans = len(pred) - len(fare["amount"])
amount = list(fare["amount"]) + [np.nan for _ in range(num_nans)]
amount = np.array(amount)

fare_result_df["amount"] = amount
fare_result_df["prediction"] = pred
fare_result_df["timestamp"] = timestamps
fare_result_df.index = fare_result_df["timestamp"]
fare_result_df.drop("timestamp", inplace=True, axis=1)
```

```python
fare_result_df.plot()
```

As you can see from the above graph, clearly we haven't made a good forecasting model.  While this model probably captures some of the trend information, it misses a lot of the other information, making it a poor forecast.

This is one of the major problems with insample prediction, things can look good, but that doesn't mean they hold up.  Clearly all the smoothing we added has lost us a lot of predictive power.  Let's try some other modeling techniques to see if we can recover a reasonable forecast.

```python
def fare_prophet_forecast(fare, changepoint=0.05):
    fare = fare.rename(columns={"timestamp": "ds", "amount": "y"})
    fare.drop("day", inplace=True, axis=1)
    fare.drop("hour", inplace=True, axis=1)
    fare_model = fbprophet.Prophet(changepoint_prior_scale=changepoint)
    fare_model.fit(fare)
    fare_forecast = fare_model.make_future_dataframe(periods=96, freq="H")
    fare_forecast = fare_model.predict(fare_forecast)
    fare_model.plot(fare_forecast, xlabel="DateTime", ylabel="Fare Amount")
    plt.title("Fare Amount Predicted 4 days ahead")
    return fare_model, fare_forecast
    
fare_model, fare_forecast = fare_prophet_forecast(fare)
```

Our prediction looks pretty good! 

```python
fare_model.plot_components(fare_forecast);
```

As you can see it appears the daily trend is such that we make the most money between 2am and 8am, which is consistent with the bump we say previously.  Also it looks like there is a clear Thursday - Sunday effect for getting more returns.  

I think it's clear from a revenue perspective to run this business on the weekends, at least in the outer boroughs.  It doesn't seem worth it to run it Monday-Wednesday.

# Forecasting Costs

Now that we have a reasonable sense of revenue, let's look at trip distance, to get a sense of our costs.

```python
def trip_prophet_forecast(trip, changepoint=0.05):
    trip = trip.rename(columns={"timestamp": "ds", "distance": "y"})
    trip.drop("day", inplace=True, axis=1)
    trip.drop("hour", inplace=True, axis=1)
    trip_model = fbprophet.Prophet(changepoint_prior_scale=changepoint)
    trip_model.fit(trip)
    trip_forecast = trip_model.make_future_dataframe(periods=96, freq="H")
    trip_forecast = trip_model.predict(trip_forecast)
    trip_model.plot(trip_forecast, xlabel="DateTime", ylabel="Fare Amount")
    plt.title("Trip Distance Predicted 4 days ahead")
    return trip_model, trip_forecast
    
trip_model, trip_forecast = trip_prophet_forecast(trip)
```

```python
trip_model.plot_components(trip_forecast);
```

It looks like trip distance increases similarly to revenue.
